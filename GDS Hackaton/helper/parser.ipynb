```python
import csv
import requests
from bs4 import BeautifulSoup
```

**Russian Language**

```python
url = "https://astana.citypass.kz/ru/category/muzei-i-galerei/"

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    
    attractions_full = soup.find_all(class_="sights__item sights__item--big")

    with open("attractions_data_ru.csv", "w", newline="", encoding="utf-8") as csvfile:
        fieldnames = ["Title", "Address", "Phone Number", "Description", "Schedule", "Image", "Link"]
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        
        for attraction in attractions_full:
            attraction_title = attraction.find(class_="sights__item--title")
            attraction_name = attraction_title.get_text(strip=True)
            attraction_link = attraction_title.find('a')['href']
            
            attraction_response = requests.get(attraction_link)
            
            if attraction_response.status_code == 200:
                attraction_soup = BeautifulSoup(attraction_response.text, 'html.parser')

                description_element = attraction_soup.find(class_="object_content--desc")
                address_element = attraction_soup.find(class_="object__info--adres")
                phone_element = attraction_soup.find(class_="object__info--phone-repeater")
                schedule_element = attraction_soup.find(class_="object_content--right-list object_content--timetable")

                image_element = attraction.find(class_="sights__item--img")
                image_src = image_element.find("img")["src"] if image_element else ""

                if description_element and address_element and phone_element and schedule_element and image_src:
                    description = description_element.get_text(strip=True)
                    address = address_element.get_text(strip=True)
                    phone_number = phone_element.find('a').get_text(strip=True)
                    
                    schedule_items = schedule_element.find_all('li')
                    schedule = []
                    for item in schedule_items:
                        day = item.find(class_="object_content-one").get_text(strip=True)
                        time = item.find(class_="object_content-too").get_text(strip=True)
                        schedule.append(f"{day}: {time}")
                    
                    print(f'Title: {attraction_name}')
                    print(f'Address: {address}')
                    print(f'Phone Number: {phone_number}')
                    print(f'Description: {description}')
                    print(f"Schedule: {schedule}")
                    print(f"Image: {image_src}")
                    print(f'Link: {attraction_link}')

                    writer.writerow({
                        "Title": attraction_name,
                        "Address": address,
                        "Phone Number": phone_number,
                        "Description": description,
                        "Schedule": ", ".join(schedule),
                        "Image": image_src,
                        "Link": attraction_link
                    })
                    
            else:
                print(f"Ошибка при получении данных с сайта {attraction_link}: {attraction_response.status_code}")
else:
    print("Ошибка при получении данных с сайта:", response.status_code)
```
